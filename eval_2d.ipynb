```python
from PIL import Image
from IPython.display import display
# 图像路径，替换为你的图像路径
image_path1 = 'exp_pic/moon_SchB.png'
image_path2 = 'exp_pic/moon_KL.png'
image_path3 = 'exp_pic/checkerboard_SchB.png'
image_path4 = 'exp_pic/checkerboard_KL.png'
def plot_2d_gen(image_path1,image_path2):
    image1 = Image.open(image_path1)
    image2 = Image.open(image_path2)
    # 确定新图像的宽度和高度
    width = max(image1.width, image2.width)
    height = image1.height + image2.height
    # 创建一个新图像，模式与原图像相同，大小为两个图像的宽度和高度之和
    new_image = Image.new('RGB', (width, height))
    # 将两个图像粘贴到新图像上，image1 在上，image2 在下
    new_image.paste(image1, (0, 0))
    new_image.paste(image2, (0, image1.height))
    # 显示结果图像
    display(new_image)
    
print("generate process" )
print("up SchB ; down KL")
plot_2d_gen(image_path1=image_path1,image_path2=image_path2)
print("generate process" )
print("up SchB ; down KL")
plot_2d_gen(image_path1=image_path3,image_path2=image_path4)
```
